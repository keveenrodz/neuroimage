# Directories path

Here you must enter the path to each directory containing the results of the Constrained Group ICA (cGICA).

This is done for each group of networks.

Example:
- G1/2_rs_networks_all
- G1/2_rs_networks_300ROI

```python
relevant_path_G1 = '/your_path/G1_rs_networks_300ROI/'
relevant_path_G2 = '/your_path/G2_rs_networks_300ROI/'

relevant_path = [relevant_path_G1, relevant_path_G2]
```

# Split 4D data

The cGICA outputs are composed of one volume for each RSN, so it is necessary to separate each RSN.

```python
for gica in relevant_path:
    files_gica = extract_gICA_files(gica)
    print(files_gica)
    for gica_name in files_gica:
        output_gica_name = gica_name[: gica_name.index('.nii')]
        split_fsl(gica + gica_name, gica + output_gica_name)
        print(f'File: {output_gica_name} DONE')
    print('_______________-')
```

# RSN Denoising

Here we take each network from each subject and by group, apply the denoising and deliver each of the respective denoising steps. This process takes a few minutes.

The denoising results have respective prefixes.

The results are delivered in the same original directory, it does NOT create a new one.

```python
for gica in relevant_path:
    files_gica = extract_gICA_files(gica)
    print(*files_gica, sep='\n')
    for gica_name in files_gica:
        print('gica name: ', gica_name)
        output_gica_name = gica_name[: gica_name.index('.nii')]
        p_val_threshold(gica, gica_name, output_gica_name)
        print(f'File: {output_gica_name} DONE')
```
